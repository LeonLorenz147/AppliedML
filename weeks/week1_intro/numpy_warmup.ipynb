# 🔢 Intro to NumPy

This notebook is a short introduction to **NumPy**, which stands for **Numerical Python**. It is a library for efficient array-based computation in Python and one of the core libraries for scientific computing and machine learning.

At the heart of NumPy is the `ndarray` type, or **NumPy array**, which represents a homogeneous array of values (usually numbers, such as integers or floats). You can think of NumPy arrays as the Python equivalent of vectors, matrices, or tensors in linear algebra. 

Unlike ordinary Python lists, NumPy arrays are designed to support **vectorized operations**, meaning that we can apply operations to whole arrays at once. This makes NumPy code not only shorter and easier to read, but also much faster in practice. Under the hood, many NumPy operations are implemented efficiently in lower-level code.

## 📦 Importing NumPy

Before using NumPy, we need to **import** the library.

Convention:  

`import numpy as np`

- `np` is just a short **alias** — we could use any name, but `np` is standard.  
- After this, every NumPy function is accessed as `np.function(...)` (for example, `np.sin`, `np.arange`).  
- This avoids confusion with other libraries that might also define functions like `sin` or `exp`. 


```python
import numpy as np 
```

## 🔢 Creating Arrays

There are many ways to create NumPy arrays:

- From Python lists: `np.array([1, 2, 3])`  
- Ranges: `np.arange(start, stop, step)` → like `range`, but returns an array  
- Evenly spaced values: `np.linspace(start, stop, num)`  
- Special arrays:  
  - `np.zeros(shape)` → all zeros  
  - `np.ones(shape)` → all ones  
  - `np.eye(n)` → identity matrix  
  - `np.full(shape, value)` → fill with a constant  

```python
# From a list
a = np.array([1, 2, 3, 4])

# Ranges
ar = np.arange(0, 10, 2)   # 0, 2, 4, 6, 8
lin = np.linspace(0, 1, 5) # 5 points from 0 to 1 inclusive

# Special arrays
zeros = np.zeros((2, 3))
ones  = np.ones((3, 3))
eye   = np.eye(3)
full  = np.full((2, 2), 7)

a, ar, lin, zeros, ones, eye, full
```

## 🎯 Indexing & Slicing

NumPy arrays use indexing rules similar to Python lists — but extended to multiple dimensions.

- **1D arrays**: `a[0]`, `a[1:4]`  
- **2D arrays**: `A[i, j]` for row `i`, column `j`  
- **Slicing**: `A[:2, 1:]` selects a submatrix  
- **Boolean masks**: select elements that satisfy a condition  
- **Fancy indexing**: select specific indices with lists or arrays


```python
A = np.arange(1, 10).reshape(3, 3)
A
```

```python
# Single element
A[0, 2]
```

```python
# Slice submatrix
A[:2, 1:]
```

```python
# Boolean mask
mask = A % 2 == 0
mask, A[mask]
```

```python
# Fancy indexing
A[[0, 2], [1, 2]]  # elements (0,1) and (2,2)
```

💡 Slicing in NumPy usually gives a **view**, not a copy. That means changing the slice also changes the original array.

```python
sub = A[:2, :2]
sub[0, 0] = 99
A   # the change appears in A too!
```

## 🧮 Array Math

NumPy arrays support both **elementwise operations** and **matrix-style operations**.

### 🔹 Elementwise
- Arithmetic: `+`, `-`, `*`, `/`, `**` → applied to each element  
- Comparisons: `>`, `<`, `==` → return boolean arrays  
- Universal functions: `np.sin`, `np.exp`, `np.log`, … applied elementwise  

### 🔹 Aggregate functions
- `np.sum`, `np.mean`, `np.max`, `np.min`, `np.std`  
- With `axis` argument: operate along rows or columns  

### 🔹 Matrix-style
- Dot product: `np.dot(a, b)` or `a @ b`  
- Matrix–vector multiplication  
- Matrix–matrix multiplication


```python
x = np.linspace(0, 2*np.pi, 5)

# Elementwise operations
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(-x)

y1, y2, y3
```

```python
# Comparisons
mask = (y1 > 5)
mask, y1[mask]
```

```python
# Aggregates
np.sum(y1), np.mean(y1), np.max(y1)
```

```python
# Matrix-style multiplication
A = np.array([[1, 2],
              [3, 4]])
b = np.array([10, 20])

# Matrix–vector
A @ b
```

```python
# Matrix–matrix
B = np.array([[2, 0],
              [1, 2]])
A @ B
```

💡 Important:
- `*` means elementwise multiplication  
- `@` or `np.dot` means matrix multiplication  


## 🔄 Array Operations

NumPy arrays can be reshaped, transposed, and combined in powerful ways.

### Reshape
- `.reshape(new_shape)` changes the dimensions of an array **without copying data**.  
- Useful for turning a 1D vector into a 2D matrix, etc.

### Transpose
- `.T` gives the transpose: rows ↔ columns.  
- Works for 2D arrays (and higher dimensions).

### Broadcasting
- NumPy automatically “stretches” smaller arrays to match shapes.  
- Lets you write **vector + scalar**, **matrix + vector**, etc. without explicit loops.


```python
# Reshape
arr = np.arange(12)
M = arr.reshape(3, 4)
M
```

```python
# Transpose
M.T
```

```python
# Broadcasting: add vector to each row
v = np.array([1, 2, 3, 4])
M + v
```

```python
# Broadcasting: scale rows differently
scales = np.array([[1], [2], [3]])  # shape (3,1)
M * scales
```

💡 Broadcasting rules:
1. Align shapes from the right.  
2. Dimensions must be equal, or one of them must be 1.  
3. NumPy “stretches” the 1 to fit.  

Example: `(3,4)` array + `(4,)` vector → `(3,4)` result.  


## 🧮 Array Math with NumPy Functions

NumPy redefines math functions (`sin`, `cos`, `exp`, …) so they work **elementwise** on arrays.


```python
x = np.linspace(0, 2*np.pi, 5)

np.sin(x), np.cos(x), np.exp(x)
```

## 🎲 Random Arrays

NumPy includes a powerful random number generator.  

### Common functions:
- `np.random.rand(m, n)` → uniform in [0, 1)  
- `np.random.randn(m, n)` → standard normal distribution (mean 0, std 1)  
- `np.random.randint(low, high, size)` → random integers  
- `np.random.seed(n)` → set the seed for reproducibility  

```python
np.random.seed(0)  # reproducibility

U = np.random.rand(2, 3)      # uniform [0,1)
N = np.random.randn(3, 3)     # normal(0,1)
R = np.random.randint(0, 10, size=(2, 5))  # random integers

U, N, R
```

## ⚡ Performance: Lists vs NumPy

Python lists are flexible, but NumPy arrays are **faster, more compact, and better suited for numerical computation**.

### Why do arrays outperform lists?

- **Homogeneous data** 
- **Vectorized operations**
- **Lower-level (C) code under the hood** 

```python
import time

# Large dataset
N = 10000000
lst = list(range(N))
arr = np.arange(N)

# Timing with Python list
start = time.time()
squares_list = [x**2 for x in lst]
end = time.time()
print("List comprehension:", end - start, "seconds")

# Timing with NumPy array
start = time.time()
squares_array = arr**2
end = time.time()
print("NumPy array:", end - start, "seconds")
```
